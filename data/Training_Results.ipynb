```python
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
import pandas as pd
```

```python
CURRENT = Path.cwd()
CURRENT
```

```python
Dennis_CNN160380_100episode_0dmp = pd.read_csv(CURRENT/'Dennis_CNN160380_100episode_0dmp/mix_DQN_0.01_0.9_32.csv')
Dennis_CNN80190_100episode_0dmp = pd.read_csv(CURRENT/'Dennis_CNN80190_100episode_0dmp/mix_DQN_0.01_0.9_32.csv')
Dennis_CNN80190LSTM_100episode_0dmp = pd.read_csv(CURRENT/'Dennis_CNN80190LSTM_100episode_0dmp/mix_DQN_0.01_0.9_32.csv')
Dennis_CNN80190LSTMA3C_100episode_0dmp = pd.read_csv(CURRENT/'Dennis_CNN80190LSTMA3C_100episode_0dmp/mix_DQN_0.01_0.9_32.csv')
```

```python
df1 = Dennis_CNN160380_100episode_0dmp.rename(columns={'reward': 'r1'})
df2 = Dennis_CNN80190_100episode_0dmp.rename(columns={'reward': 'r2'})
df3 = Dennis_CNN80190LSTM_100episode_0dmp.rename(columns={'reward': 'r3'})
df4 = Dennis_CNN80190LSTMA3C_100episode_0dmp.rename(columns={'reward': 'r4'})
```

```python
result = pd.concat([df1.r1, df2.r2, df3.r3, df4.r4], axis=1, sort=False)
```

```python
res = result.rolling(10).mean()
res
```

```python
plt.plot(res.index, res.r1)
plt.plot(res.index, res.r2)
plt.plot(res.index, res.r3)
plt.plot(res.index, res.r4)

plt.legend(['CNN160x380', 'CNN80x190', 'CNN80x190_LSTM', 'CNN80x190_LSTM_A3C'])
plt.title("Training Reward")
plt.xlabel("Episode")
plt.ylabel("Reward")
```

```python
result['CNN160x380'] = result['r1']
result['CNN80x190'] = result['r2']
result['CNN80x190_LSTM'] = result['r3']
result['CNN80x190_LSTM_A3C'] = result['r4']
result.head()
```

```python
print("Average Reward over 100 episode")
f = result[['CNN160x380', 'CNN80x190', 'CNN80x190_LSTM', 'CNN80x190_LSTM_A3C']].mean()
f
```
